```python
import pandas as pd
import numpy as np
import datetime as dt
from pathlib import Path
%matplotlib inline
```

## Data Cleaning

Read the Portfolio daily returns and clean the data

```python
# Reading whale returns
whale_returns_csv = Path("Resources/whale_returns.csv")

# my notes - use pd.read_csv and parameters parse_dates and infer_datetime_format to index on the Date column
whale_returns_df = pd.read_csv(whale_returns_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
whale_returns_df.sort_index(ascending=True, inplace=True)

# my notes - use .head() to look at the dataframe
whale_returns_df.head()
```

```python
# Count nulls
# my notes - use .isnull().sum()
whale_returns_df.isnull().sum()
```

```python
# Drop nulls
# my notes - use .dropna()
whale_returns_df = whale_returns_df.dropna()
whale_returns_df.isnull().sum()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns_csv = Path("Resources/algo_returns.csv")

# my notes - use pd.read_csv and parameters parse_dates and infer_datetime_format to index on the Date column
algo_returns_df = pd.read_csv(algo_returns_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
algo_returns_df.sort_index(ascending=True, inplace=True)

# my notes - use .head() to look at the dataframe
algo_returns_df.head()
```

```python
# Count nulls
# my notes - use .isnull().sum()
algo_returns_df.isnull().sum()
```

```python
# Drop nulls
# my notes - use .dropna()
algo_returns_df = algo_returns_df.dropna()
algo_returns_df.isnull().sum()
```

## S&P 500 Returns

Read the S&P500 Historic Closing Prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500_history_csv = Path("Resources/sp500_history.csv")

# my notes - use pd.read_csv and parameters parse_dates and infer_datetime_format to index on the Date column
sp500_history_df = pd.read_csv(sp500_history_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
sp500_history_df.sort_index(ascending=True, inplace=True)

# my notes - use .head() to look at the dataframe
sp500_history_df.head()
```

```python
# Check Data Types
# my notes - use .dtypes
sp500_history_df.dtypes
```

```python
# Fix Data Types
# my notes - use .astype() to change data type. (THIS DID NOT WORK, had to research solution)
sp500_history_df['Close'] = sp500_history_df['Close'].str.replace('$', '')
sp500_history_df['Close'] = pd.to_numeric(sp500_history_df['Close'])
sp500_history_df.dtypes
```

```python
# Calculate Daily Returns
# my notes - use set dataframe name sp500_daily_returns and use .pct_change() and .head() to view the dataframe
sp500_daily_returns = sp500_history_df.pct_change()
sp500_daily_returns.head()
```

```python
# Drop nulls
# my notes - use .dropnna
sp500_daily_returns = sp500_daily_returns.dropna()
sp500_daily_returns.head()
```

```python
# Rename Column
# my notes - use .columns = ['new column name'] to change Close to SP500
sp500_daily_returns.columns = ['S&P 500']
sp500_daily_returns.head()
```

## Combine Portfolios, Algorithmic, and S&P 500 Returns

```python
# Concatenate all DataFrames into a single DataFrame
# my notes - use pd.concate to combine on columns using inner join. Check dates for each dataframe first to confirm this is the correct method to use
combined_daily_returns = pd.concat([whale_returns_df, algo_returns_df, sp500_daily_returns], axis='columns', join='inner')
combined_daily_returns.head()
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Plot the rolling standard deviation of the various portfolios along with the rolling standard deviation of the S&P 500 (consider a 21 day window). Does the risk increase for each of the portfolios at the same time risk increases in the S&P?
    - Response: Yes, the risk increases for each portofolio at the same time the risk increases in the S&P.  Although there are two times, at the beginning of 2017 and 2019, where Tiger Global Management's risk increases as the S&P's risk was declining.
2. Construct a correlation table for the algorithmic, whale, and S&P 500 returns. Which returns most closely mimic the S&P?
    - Response: With a correlation of 0.858764, Algo 2 most closely mimic the S&P. The Soros Fund Management LLC portfolio is another portfolio that closesly mimics the S&P with a correlation of 0.837864.
3. Choose one portfolio and plot a rolling beta between that portfolio's returns and S&P 500 returns. Does the portfolio seem sensitive to movements in the S&P 500?
    - Response: Overall the Tiger Global Management LLC portfolio does not seem to be sensitive to the movements in the S&P 500.  Although there were periods in 2016 and 2018 where it was more sensative to the S&P movements.
4. An alternative way to calculate a rolling window is to take the exponentially weighted moving average. This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the ewm with a 21 day half-life.

```python
# Calculate and plot the rolling standard deviation for
# the S&P 500 and whale portfolios using a 21 trading day window
# my notes - use .rolling(window=21).std() for both SP500 and whale portfolios
# my notes - use .plot() to show the graph. use the title parameter to add a title and figsize=(20,10) to make the graph larger
rolling_std = combined_daily_returns.rolling(window=21).std()
rolling_std.plot(figsize=(20,10), title="21 Day Rolling Standard Devations")
```

```python
# Construct a correlation table
# my notes - use .corr() on the combined dataframe
correlation = combined_daily_returns.corr()
correlation
```

```python
# Calculate Beta for a single portfolio compared to the total market (S&P 500)
# my notes - select one portfolio and use the beta formula (covariance / variance) to compare to the SP500
# my notes - use .plot to show the graph. use title parameter to add a title and figsize=(20,10) to make the chart larger
rolling_covariance = combined_daily_returns["TIGER GLOBAL MANAGEMENT LLC"].rolling(window=21).cov(combined_daily_returns['S&P 500'])
rolling_variance = combined_daily_returns['S&P 500'].rolling(window=21).var()
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(20,10), title="Tiger Global Management LLC Rolling Beta")
```

```python
# Calculate a rolling window using the exponentially weighted moving average. 
# my notes - research the ewa formula (after research use .ewma() function)
# my notes - use .rolling().mean() with the window parameter(unknown as of this note)
# my notes - use .plot() to show the graph. use title parameter to add the title and figsize=(20,10) to make the chart larger 
combined_exp_wma = combined_daily_returns.ewm(halflife=21).mean()
combined_exp_wma.plot(figsize=(20,10), title='Exponential Weighted Average')
```

## Sharpe Ratios

1. Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot.
2. Determine whether the algorithmic strategies outperform both the market (S&P 500) and the whales portfolios.

```python
# Calculate annualized Sharpe Ratios
# my notes - set to sharpe ratio variable name and use formula ((all portfolio returns.mean() * 252 / (all portfolio returns.std() * np.sqrt(252)))
sharpe_ratios = (combined_daily_returns.mean() * 252 / (combined_daily_returns.std() * np.sqrt(252)))
sharpe_ratios
```

```python
# Visualize the sharpe ratios as a bar plot
# my notes - use .plot() to show the graph. use the parameter kind and set it bar and title parameter to add a title
sharpe_ratios.plot(kind='bar', title="Sharpe Ratios")
```

On the basis of this performance metric, do our algo strategies outperform both 'the market' and the whales? Type your answer here:
- Response: Base on the performance metrics, only Algo 1 outperforms both "the market" and the whale portfolios.  Although Algo 2 does well it doesn't outperform "the market" or Berkshire Hathaway.

```python
# Read the first stock
fb_csv = Path('Resources/fb_google_finance.csv')
fb_df = pd.read_csv(fb_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
fb_df.sort_index(ascending=True, inplace=True)
fb_df.head()
```

```python
# Read the second stock
cvx_csv = Path('Resources/cvx_google_finance.csv')
cvx_df = pd.read_csv(cvx_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
cvx_df.sort_index(ascending=True, inplace=True)
cvx_df.head()
```

```python
# Read the third stock
fisv_csv = Path('Resources/fisv_google_finance.csv')
fisv_df = pd.read_csv(fisv_csv, index_col='Date', parse_dates=True, infer_datetime_format=True)
fisv_df.sort_index(ascending=True, inplace=True)
fisv_df.head()
```

```python
# Concatenate all stocks into a single DataFrame
brett_portfolio = pd.concat([fb_df, cvx_df, fisv_df], axis='rows', join='inner')
brett_portfolio.head()
```

```python
# Reset the index
brett_portfolio.reset_index().head()
```

```python
# Pivot so that each column of prices represents a unique symbol
brett_portfolio = brett_portfolio.pivot(columns='Symbol', values='Close')
brett_portfolio.head()
```

```python
# Drop Nulls
brett_returns = brett_portfolio.pct_change().dropna()
brett_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Calculate weighted portfolio returns
weights = [1/3, 1/3, 1/3]
b_porfolio_returns = brett_returns.dot(weights)
b_porfolio_returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Add your "Custom" portfolio to the larger dataframe of fund returns
all_daily_returns = pd.concat([combined_daily_returns, b_porfolio_returns], axis='columns', join='inner')
all_daily_returns.rename(columns={0:'Brett Portfolio'}, inplace=True)
all_daily_returns
```
